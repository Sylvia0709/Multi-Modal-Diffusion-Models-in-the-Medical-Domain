```python
import shutil, os
src_dir1 = "results/KIRC-CT2MR/test_70/images"
src_dir2 = "results/KIRC-MR2CT/test_70/images"
src_dir3 = "results/LIHC-CT2MR/test_100/images"
src_dir4 = "results/LIHC-MR2CT/test_100/images"
dst_dir1 = "RSummary/KIRC-CT2MR"
dst_dir2 = "RSummary/KIRC-MR2CT"
dst_dir3 = "RSummary/LIHC-CT2MR"
dst_dir4 = "RSummary/LIHC-MR2CT"
os.makedirs(dst_dir1, exist_ok=True)
os.makedirs(dst_dir2, exist_ok=True)
os.makedirs(dst_dir3, exist_ok=True)
os.makedirs(dst_dir4, exist_ok=True)
for fname in os.listdir(src_dir1):
    if "fake_B" in fname or "real_A" in fname or "real_B" in fname:
        shutil.move(os.path.join(src_dir1, fname), os.path.join(dst_dir1, fname))
for fname in os.listdir(src_dir2):
    if "fake_A" in fname or "real_A" in fname or "real_B" in fname:
        shutil.move(os.path.join(src_dir2, fname), os.path.join(dst_dir2, fname))
for fname in os.listdir(src_dir3):
    if "fake_B" in fname or "real_A" in fname or "real_B" in fname:
        shutil.move(os.path.join(src_dir3, fname), os.path.join(dst_dir3, fname))
for fname in os.listdir(src_dir4):
    if "fake_A" in fname or "real_A" in fname or "real_B" in fname:
        shutil.move(os.path.join(src_dir4, fname), os.path.join(dst_dir4, fname))
```
